```python
required_columns = ["BedroomAbvGr" , "HalfBath" , "FullBath" , "SalePrice"]
```

## PipeLine

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = required_dataset.drop("SalePrice" ,axis=1)
Y= required_dataset["SalePrice"].copy()
X
# Y
```

```python
from sklearn.pipeline import Pipeline
data_pipeline = Pipeline([
    ("std",StandardScaler())
])
```

```python
tr_x = data_pipeline.fit_transform(X)
```

```python
tr_x
tr_x_df = pd.DataFrame(tr_x , columns= X.columns)
tr_x_df
```

## train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(tr_x_df,Y ,test_size=0.2 , random_state=42)
x_train
```

## model building

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
predictions = model.predict(x_train)
pred_df = pd.DataFrame(predictions)
pred_df
```
